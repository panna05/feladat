```sql
SELECT 
MIN(listaar) AS 'minimum_listaar',
MAX(listaar) AS 'maximum_listaar', 
AVG(listaar) AS 'átlag_listaar'   
FROM Termek
WHERE listaar IS NOT NULL 
```

```sql
SELECT 
    COUNT(SZULO_KAT) AS 'Nem üres',
    COUNT(*)-COUNT(SZULO_KAT) AS 'Üres sorok',
    CAST(COUNT(SZULO_KAT)* 100.0  / COUNT(*) AS DECIMAL(5,2))  as 'Nem üres %',
    CAST((COUNT(*)-COUNT(SZULO_KAT))* 100.0  / COUNT(*) AS DECIMAL(5,2))  as 'Üres %'
FROM Termekkategoria
```

```sql
SELECT 
    felvitte, 
    COUNT(termekkod)
from Termek
WHERE felvitte IS NOT null
GROUP BY felvitte
ORDER BY Count(termekkod)
```

```sql
SELECT LEN(MEGNEVEZES) AS 'szöveg_hossz',count(*) as 'Előfordulás'
FROM Termek
GROUP BY LEN(MEGNEVEZES)
ORDER BY 1 DESC
```
