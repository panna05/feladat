```sql
-- 1.feladat
SELECT TIPUS, AVG(CSILLAGOK_SZAMA) as 'Átlag'
FROM Szallashely
where TIPUS <> 'panzió'
group by tipus
having AVG(CSILLAGOK_SZAMA) < 3
```

```sql
--2.feladat
SELECT sz.SZOBA_ID,sz.SZALLAS_FK,sz.SZOBA_SZAMA, sz.FEROHELY,sz.POTAGY,sz.KLIMAS, h.SZALLAS_NEV, 
case
when h.TIPUS = 'panzió' then 'Igen'
else 'Nem'
end as 'Panzió'
FROM Szoba sz join Foglalas f on sz.SZOBA_ID=f.SZOBA_FK
                join Szallashely h on sz.SZALLAS_FK=h.SZALLAS_ID




```
